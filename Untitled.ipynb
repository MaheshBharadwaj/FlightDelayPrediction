```python
import numpy as np
import pandas as pd
import pickle
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pickle.load(open('./JoinedData.pkl','rb'))
```

```python
df.dropna(how='any',inplace=True)
```

```python
fig,ax = plt.subplots(figsize=(10,7))

ax.hist(df[df['ArrDel15'] == 1]['ArrDel15'],color='r',label = 'Delayed')
ax.hist(df[df['ArrDel15'] == 0]['ArrDel15'],color='b',label = 'On Time')
ax.set_xlabel('Class',fontsize=12)
ax.set_ylabel('Frequency',fontsize=12)
ax.set_title('Flights in Data Set',fontsize=15)
ax.legend(loc='upper right',fontsize=15)
plt.savefig('./Images/DatasetClassDistributionHistogram.png')

plt.show()
```
